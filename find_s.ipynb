```python
import pandas as pd
import numpy as np
data = pd.read_csv("Wholesale_customers.csv")
data['Grocery'] = np.where(
    data['Grocery'] >= data['Grocery'].median(), 'High', 'Low'
)
data['Detergents_Paper'] = np.where(
    data['Detergents_Paper'] >= data['Detergents_Paper'].median(), 'High', 'Low'
)
data['Channel'] = data['Channel'].map({1: 'Horeca', 2: 'Retail'})
data['Region'] = data['Region'].map({
    1: 'Region1',
    2: 'Region2',
    3: 'Region3'
})
data['Class'] = np.where(
    (data['Grocery'] == 'High') &
    (data['Detergents_Paper'] == 'High'),
    'Yes', 'No'
)
X = data[['Channel', 'Region', 'Grocery', 'Detergents_Paper']]
y = data['Class']
hypothesis = None
for i in range(len(y)):
    if y.iloc[i] == 'Yes':
        hypothesis = list(X.iloc[i])
        break
for i in range(len(y)):
    if y.iloc[i] == 'Yes':
        for j in range(len(hypothesis)):
            if X.iloc[i, j] != hypothesis[j]:
                hypothesis[j] = '?'
print("Final Hypothesis:", hypothesis)
```
