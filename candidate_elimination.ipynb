```python
import pandas as pd
data = pd.read_csv("Wholesale_customers.csv")
features = data[['Fresh', 'Milk', 'Grocery', 'Frozen']]
target_raw = data['Channel']
def convert_to_category(column):
    mean_val = column.mean()
    return ['High' if val >= mean_val else 'Low' for val in column]
concepts = pd.DataFrame({
    'Fresh': convert_to_category(features['Fresh']),
    'Milk': convert_to_category(features['Milk']),
    'Grocery': convert_to_category(features['Grocery']),
    'Frozen': convert_to_category(features['Frozen'])
}).values.tolist()
target = ['Yes' if val == 1 else 'No' for val in target_raw]
def candidate_elimination(concepts, target):
    specific_h = concepts[0].copy()
    general_h = [['?' for _ in range(len(specific_h))]
                 for _ in range(len(specific_h))]
    for i, instance in enumerate(concepts):
        if target[i] == 'Yes':
            for x in range(len(specific_h)):
                if instance[x] != specific_h[x]:
                    specific_h[x] = '?'
                    general_h[x][x] = '?'
        elif target[i] == 'No':
            for x in range(len(specific_h)):
                if instance[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = '?'
    general_h = [g for g in general_h if g != ['?', '?', '?', '?']]
    return specific_h, general_h
S_final, G_final = candidate_elimination(concepts, target)
print("Final Specific Boundary (S):", S_final)
print("Final General Boundary (G):", G_final)
```
